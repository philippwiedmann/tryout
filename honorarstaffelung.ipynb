```python
max_an = [3, 10, 20, 30, 50, 200]
target_price = [25, 21, 18.5, 16.5, 14.5, 12.5]
```

```python
min_an = [1] + [i + 1 for i in max_an[:-1]] 
price = 6*[0]

price[0] = target_price[0]
for i in range(1, len(max_an)):
    price[i] = (sum(k for k in range(max_an[i-1] + 1, max_an[i] + 1)) * target_price[i] - (max_an[i] - max_an[i-1]) * sum(price[j] * (max_an[j] - (min_an[j] - 1)) for j in range(0, i))) / sum(k for k in range(1,  max_an[i] - max_an[i-1] + 1))
print(price)
```

```python
actual_price = price
```

```python
x = list(range(0, 201))
y = 201*[0]
for i in range(0, len(max_an)):
    for j in range(min_an[i], max_an[i] + 1):
        y[j] = actual_price[i] * (j-(min_an[i] - 1)) + sum(price[k] * (max_an[k] - (min_an[k] - 1)) for k in range(0, i))

import matplotlib.pyplot as plt
plt.plot(x, y)
```
